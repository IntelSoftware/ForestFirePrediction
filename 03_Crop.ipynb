# Trim images to be same size

Necessary to thumbnail down the image becuse the raw or even slightly smaller cropped images cause resnet to have to allocate too much memory ~ 59 GB

```python
from PIL import Image
import os

import os 
try:
    os.makedirs('data/CA_Fire/NoFire')
    os.makedirs('data/CA_Fire/Fire')
except:
    print("folder already exist")

#Opens the image

def cropIm(fldrIn, fn, fldrOut, type='PNG'):
    path = fldrIn + fn
    im = Image.open(path)
    rgb_im = im.convert('RGB')
    #Specifies the corners of the desired image
    # left = 95
    # top = 98
    # right = 3752
    # bottom = 2281

    left = 150
    top = 150
    right = 150 + 224*16
    bottom = 150 + 224*9

    #Crops the image
    im = rgb_im.crop((left, top, right, bottom))
    
    # resize - otherwise renet cant allocate enough memory
    MAX_SIZE = (1024, 1024)
    rgb_im.thumbnail(MAX_SIZE)

    print(im.size, fn)
    #Saves the image
    rgb_im.save(fldrOut + fn[:-4] + f'.{type}', format=type)
    rgb_im.close()
# cropIm(fldrIn, fn, fldrOut)

import glob
FireList = []
NoFireList = []

fldrIn = "CA_Fire/Fire/"
fldrOut = "data/CA_Fire/Fire/"
for fire, fn in enumerate(glob.glob(fldrIn+"*.tif")):
    fn_trunc = fn.split('/')[-1]
    FireList.append(fn_trunc)
    cropIm(fldrIn, fn_trunc, fldrOut, 'PNG')
    
print('Count fire files', fire)

fldrIn = "CA_Fire/NoFire/"
fldrOut = "data/CA_Fire/NoFire/"
for noFire, fn in enumerate(glob.glob(fldrIn+"*.tif")):
    fn_trunc = fn.split('/')[-1]
    NoFireList.append(fn_trunc)
    cropIm(fldrIn, fn_trunc, fldrOut, 'PNG')
print('Count NoFire files', noFire)
```
